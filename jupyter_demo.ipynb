```python
from PIL import Image
import matplotlib.pyplot as plt

def read_images(background_path, foreground_path):
    background = Image.open(background_path)
    foreground = Image.open(foreground_path)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Background Image")
    plt.imshow(background)
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Foreground Image")
    plt.imshow(foreground)
    plt.axis('off')

    plt.show()

# define paths
background_path = '/home/ubuntu/ghost/background_image.png'
foreground_path = '/home/ubuntu/ghost/foreground_image.png'

# call function
read_images(background_path, foreground_path)
```
